```python
import seaborn.objects as so
import seaborn as sns

df = (
    sns.load_dataset('anscombe')
    .assign(index=lambda x: x.groupby('dataset').cumcount()+1)
)

print(df)
```

```python
(
    so.Plot(df, x='x', y='y')
    .add(so.Dots())
    .facet('dataset', wrap=2)
)
```

```python
(
    so.Plot(df, x='x', y='y', color='dataset')
    .add(so.Dots())
)
```

```python
(
    so.Plot(df, x='x', y='y', marker='dataset')
    .add(so.Dots())
)
```

```python
(
    so.Plot(df, x='x', y='y', pointsize='dataset')
    .add(so.Dots())
)
```

```python
(
    so.Plot(df, x='x', y='y', 
            color='dataset', 
            marker='dataset', 
            pointsize='dataset')
    .add(so.Dots())
)
```

```python
(
    so.Plot(df, x='x', y='y', 
            color='dataset', 
            marker='dataset', 
            pointsize='dataset')
    .add(so.Dots())
    .facet('dataset', wrap=2)
)
```

```python
(
    so.Plot(df, x='x', y='y', color='index')
    .add(so.Dot())
    .facet('dataset', wrap=2)
)
```

```python
(
    so.Plot(df, x='x', y='y')
    .add(so.Path())
    .add(so.Dot(), color='index')
    .facet('dataset', wrap=2)
)
```

```python
(
    so.Plot(df, 
            x='dataset', 
            y='index', 
            color='x', 
            pointsize='y')
    .add(so.Dots())
)
```

```python
(
    df
    .melt(id_vars=['dataset', 'index'], 
          value_vars=['x', 'y'], 
          var_name='coordinate')
    .pipe(so.Plot, x='coordinate', y='index', pointsize='value')
    .add(so.Dots())
    .facet('dataset')
    .label(x=None)
)
```

```python
(
    df
    .melt(id_vars=['dataset', 'index'], 
          value_vars=['x', 'y'], 
          var_name='coordinate')
    .pipe(so.Plot, x='value', y='index')
    .add(so.Bar())
    .facet(col='dataset', row='coordinate')
    .scale(y=so.Nominal())
)
```

```python
(
    df
    .melt(id_vars=['dataset', 'index'], 
          value_vars=['x', 'y'], 
          var_name='coordinate')
    .pipe(so.Plot, x='value', y='index')
    .add(so.Line(marker='o'))
    .facet(col='dataset', row='coordinate')
    .scale(y=so.Nominal())
)
```

```python
(
    df
    .melt(id_vars=['dataset', 'index'], 
          value_vars=['x', 'y'], 
          var_name='coordinate')
    .pipe(so.Plot, x='coordinate', y='value', group='index')
    .add(so.Line(marker='o'))
    .facet('dataset')
    .label(x=None)
)
```
